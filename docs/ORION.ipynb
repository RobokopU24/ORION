## Developer Setup - Building a Graph from Source

For developers who want to clone the repository and build graphs from scratch:

### 1. Create a parent directory and clone the repository

```python
%%bash
mkdir ~/ORION_root
cd ~/ORION_root
git clone https://github.com/RobokopU24/ORION.git
```

### 2. Set up the environment

Run the setup script to create the necessary directories and set environment variables:

```python
%%bash
cd ~/ORION_root/ORION/
source ./set_up_test_env.sh
```

### 3. Build the Example_Graph

Run the build_manager to build the Example_Graph:

```python
%%bash
cd ~/ORION_root/ORION/
docker compose run --rm orion python /ORION/Common/build_manager.py Example_Graph
```
